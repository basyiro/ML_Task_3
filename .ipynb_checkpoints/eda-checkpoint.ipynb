## Looking at missing/impossible numerical information
We first check if there is any data lesser than 0 (which should not be a possible value)

```python
for numerical_feature in numerical_feature_list:
    check_less_than_zero_df = patients_df.loc[
        patients_df[numerical_feature] < 0
    ]
    if len(check_less_than_zero_df) > 0:
        print(f"{numerical_feature} has {len(check_less_than_zero_df)} negative values")
```

Thus far, there seems to be errors in __Age__ (values < 0) and __Creatinine__ (missing values). We will rectify the values for __Age__

```python
plt.scatter(patients_df.index, patients_df['Age'])
```

From the above plot, it is likely that during data entry, the age was accidentally keyed in as negative. Hence, we replace the negative ages with their positive counterparts. There also doesn't seem to be any outlier ages on the lower bound (eg, very young patients)

```python
patients_df["Age"] = patients_df["Age"].abs()
```

We will now look at how to recitfy the values for __Creatinine__. Let's check how the missing values in __Creatinine__ affects the __Survive__

```python
patients_df_copy = patients_df.copy()
patients_df_copy = patients_df_copy[patients_df_copy["Creatinine"].isnull()]
```

```python
ax = patients_df_copy["Survive"].value_counts().plot(kind="bar")
```

From the above bar chart, we see that everyone who's value is missing from __Creatinine__ did not survive the disease. Hence, we could fill in the missing values by deriving from the non-survivors' __Creatinine__ value. Let's fill up the __Creatinine__ value with the median value from non-survivors.

```python
patients_df_copy = patients_df
survived_df = patients_df_copy[patients_df_copy["Survive"] == 1]
not_survive_df = patients_df_copy[patients_df_copy["Survive"] == 0]
```

```python
imp_creatinine = SimpleImputer(missing_values=np.nan, strategy='median')
imp_creatinine.fit(not_survive_df['Creatinine'].values.reshape(-1,1))
SimpleImputer()
patients_df['Creatinine'] = imp_creatinine.fit_transform(patients_df[['Creatinine']]).ravel()
patients_df['Creatinine'].head()
```

```python
# Check how many unique values per feature for the overall, survived and not_survive
for feature in numerical_feature_list:
    overall_len = len(patients_df[feature].value_counts().sort_index())
    survived_len = len(survived_df[feature].value_counts().sort_index())
    not_survive_len = len(not_survive_df[feature].value_counts().sort_index())
    print(
        f"{feature} has {overall_len} unique values in total, with {survived_len} unique values for survived and {not_survive_len} unique values for not survive entries"
    )
```

```python
# To accomodate for the difference in length, we fill the missing unique values for survive/ not_survive
# from the overall unique values.

def make_same_length(feature):
    overall = patients_df[feature].value_counts().sort_index()
    overall_copy = overall.copy(deep=True) # Copy is set to 0 and combined with survived and not_survive
    survived = survived_df[feature].value_counts().sort_index()
    not_survive = not_survive_df[feature].value_counts().sort_index()
    
    for entry in range(len(overall_copy)):
        overall_copy.iloc[entry] = 0

    survived = survived.combine(overall_copy, max, fill_value = 0)
    not_survive = not_survive.combine(overall_copy, max, fill_value = 0)
    
    return survived, not_survive, overall
```

```python
# Compare the survival rate of the feature at each value against the overall population
def survival_rate(feature):
    survived_series, not_survive_series, overall_series = make_same_length(feature)
    
    # Convert the values to percentages for easier comparison
    for count_index in range(len(overall_series)):
        survived_series.iloc[count_index] = survived_series.iloc[count_index]/overall_series.iloc[count_index]
        not_survive_series.iloc[count_index] = not_survive_series.iloc[count_index]/overall_series.iloc[count_index]
        
    return survived_series, not_survive_series
```

We now compare the different features to see how they contribute to the survival rate.
1. The first graph shows the distribution of of the survivors vs non-survivors over the total data
2. The second graph shows the survival rate across the (increasing) values of a feature.

```python
fig, axs = plt.subplots(ncols=2, nrows=5, figsize=(20, 20))
plt.subplots_adjust(right=1.1, top=5)
graph_counter = 1

for feature in numerical_feature_list:
    plt.subplot(10, 2, graph_counter)
    patients_df[feature].value_counts().sort_index().plot(kind="line")
    survived_df[feature].value_counts().sort_index().plot(kind="line")
    not_survive_df[feature].value_counts().sort_index().plot(kind="line")
    plt.legend(
        ["Overall", "Survived", "Did not survive"],
        loc="upper center",
        prop={"size": 18},
    )

    plt.xlabel("{}".format(feature), size=20, labelpad=15)
    plt.ylabel("Patient Count", size=20, labelpad=15)
    plt.tick_params(axis="x", labelsize=20)
    plt.tick_params(axis="y", labelsize=20)
    plt.title(f"Distribution for {feature}", size=20, y=1.05)
    graph_counter += 1

    plt.subplot(10, 2, graph_counter)
    survived_series, not_survive_series = survival_rate(feature)
    survived_series.plot(kind="line", label="survive")

    plt.xlabel("{}".format(feature), size=20, labelpad=15)
    plt.ylabel("Survival Rate", size=20, labelpad=15)
    plt.tick_params(axis="x", labelsize=20)
    plt.tick_params(axis="y", labelsize=20)

    plt.legend(["Survived"], loc="upper center", prop={"size": 18})
    plt.title(f"Survival rate for {feature}", size=20, y=1.05)
    graph_counter += 1

plt.show()
```

From the above, some major features with clear trends that could've contributed to survival rates are:
1. __Hemoglobin__ (lower hemoglobin tends to survive)
2. __Weight__ (higher weight tends to survive)

For the graphs above, ignore regions where there is very little data (eg for __Creatinine__, we look at the region between 0 to 2)

For __Sodium__, __Creatinine__, __Platelets__ and __Creatinine phosphokinase__, we may need to apply other methods to visualize properly due to the skewed values.

# 2. Feature Engineering
We will now check if there are better ways of representing the features to deal with outliers, and also encode the features. We will analyze the numerical features then categorical features.

## 2.1 Binning
We can now try binning very skewed data like __Sodium__, __Creatinine__, __Platelets__ and __Creatine phosphokinase__.

```python
binned_features = ["Sodium", "Creatinine", "Platelets", "Creatine phosphokinase"]

for feature in binned_features:
    patients_df[feature] = pd.qcut(patients_df[feature], 5)
```

```python
for feature in binned_features:
    fig, axs = plt.subplots(figsize=(22, 9))
    sns.countplot(x=feature, hue='Survive', data=patients_df)

    plt.xlabel(feature, size=15, labelpad=20)
    plt.ylabel('Patient Count', size=15, labelpad=20)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=15)

    plt.legend(['Not Survived', 'Survived'], loc='upper right', prop={'size': 15})
    plt.title(f'Count of Survival in {feature}', size=15, y=1.05)

    plt.show()
```

From the above, we observe the following for the skewed features:
1. Lower __Sodium__ has higher survival rates.
2. Higher __Creatinine__ has higher survival rates
3. The spread for __Platelets__ and __Creatinine phosphokinase__ are rather well distributed with no clear relationships, but for __Platelet__ counts between 265K-319K and for __Creatinine phosphokinase__ between 22-101, the survival rates are lower.

```python
combined_ejection_fraction_mapping = {"Normal": "Normal/High", "High": "Normal/High"}
patients_df = patients_df.replace({"Ejection Fraction": combined_ejection_fraction_mapping})
survived_df = survived_df.replace({"Ejection Fraction": combined_ejection_fraction_mapping})
not_survive_df = not_survive_df.replace({"Ejection Fraction": combined_ejection_fraction_mapping})
```

# 3 Models

We define 3 models, knn neural network classifier, random forest classifier and adaboost classifier. We create pipelines for these models.

```python
model_list = []

# K-Nearest Neighbors Classifier
knn_clf = KNeighborsClassifier()

# Random Forest Classifier 
rf_clf = RandomForestClassifier(criterion='gini',
                               n_estimators=1750,
                               max_depth=7,
                               min_samples_split=6,
                               min_samples_leaf=6,
                               max_features="auto",
                               oob_score=True,
                               random_state=13,
                               n_jobs=-1,
                               verbose=1)

# Neural Network Classifier
mlp_clf = MLPClassifier(solver='adam',
                       batch_size=1000,
                       alpha=1e-5,
                       max_iter=500,
                       early_stopping=False,
                       verbose=True,
                       hidden_layer_sizes=(50, 50),
                       random_state=100)
```

We split our dataset into training and testing using `train_test_split` then use `StandardScaler` to standardize our features for each of the dataset.

```python
# Steps for pipeline
knn_clf_steps = [('scaler', StandardScaler()), ('knn_clf', KNeighborsClassifier())]
rf_clf_steps = [('scaler', StandardScaler()), ('rf_clf', RandomForestClassifier())]
mlp_clf_steps = [('scaler', StandardScaler()), ('mlp_clf', MLPClassifier())]

# Instantiate the Pipeline
knn_pipeline = Pipeline(knn_clf_steps)
rf_pipeline = Pipeline(rf_clf_steps)
mlp_pipeline = Pipeline(mlp_clf_steps)

# Define parameters for Grid Search Cross Validation
knn_parameters = {'knn_clf__n_neighbors': [5,10,15,20,25,30,35,40,45,50]}
knn_cv = GridSearchCV(knn_pipeline, param_grid=knn_parameters)

rf_parameters = {'rf_clf__n_estimators': np.arange(250, 2000, 250)}
rf_cv = GridSearchCV(rf_pipeline, param_grid=rf_parameters)

mlp_parameters = {'mlp_clf__hidden_layer_sizes': (10,10)}
mlp_cv = GridSearchCV(mlp_pipeline, param_grid=mlp_parameters)
```

```python
# Train test split
X = patients_df_encoded.drop(['Survive', 'ID'], axis=1).values
y = patients_df_encoded['Survive'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
# KNN Model Fitting

knn_cv.fit(X_train, y_train)
y_pred = knn_cv.predict(X_test)

print("KNN Accuracy: {:.2f}%".format(knn_cv.score(X_test, y_test)*100))
print("KNN F1 Score: {:.2f}".format(f1_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
print("Best KNN Parameters: {}".format(knn_cv.best_params_))
```

```python
# Random Forest Model Fitting

rf_cv.fit(X_train, y_train)
y_pred = rf_cv.predict(X_test)

print("RF Accuracy: {:.2f}%".format(rf_cv.score(X_test, y_test)*100))
print("RF F1 Score: {:.2f}".format(f1_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
print("Best RF Parameters: {}".format(rf_cv.best_params_))
```

```python
# MLP Model Fitting

mlp_cv.fit(X_train, y_train)
y_pred = mlp_cv.predict(X_test)

print("MLP Accuracy: {:.2f}%".format(mlp_cv.score(X_test, y_test)*100))
print("MLP F1 Score: {:.2f}".format(f1_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
print("Best MLP Parameters: {}".format(mlp_cv.best_params_))
```
